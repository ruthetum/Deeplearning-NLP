# 다중 선형 회귀

```
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras import optimizers

X=np.array([[70,85,11],[71,89,18],[50,80,20],[99,20,10],[50,10,10]]) # 입력 차원 3
y=np.array([73,82,72,57,34]) # 출력 차원 1

model=Sequential()
model.add(Dense(1, input_dim=3, activation='linear'))
sgd=optimizers.SGD(lr=0.00001)
model.compile(optimizer=sgd ,loss='mse',metrics=['mse']) # sgd 경사하강법, 손실함수 MSE
model.fit(X,y, batch_size=1, epochs=2000, shuffle=False) # 에포크 2000
```

```
print(model.predict(X))
```
