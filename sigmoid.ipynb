# Sigmoid 함수

```
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

W(가중치)가 커짐에 따라 경사가 커지고

W가 작아짐에 따라 경사가 작아짐

```
def sigmoid(x):
    return 1/(1+np.exp(-x))
x = np.arange(-5.0, 5.0, 0.1)
y1 = sigmoid(0.5*x)
y2 = sigmoid(x)
y3 = sigmoid(2*x)

plt.plot(x, y1, 'r', linestyle='--') # W의 값이 0.5일때
plt.plot(x, y2, 'g') # W의 값이 1일때
plt.plot(x, y3, 'b', linestyle='--') # W의 값이 2일때
plt.plot([0,0],[1.0,0.0], ':') # 가운데 점선 추가
plt.title('Sigmoid Function')
plt.show()
```

b(편향)의 값에 따라서도 그래프가 이동

```
def sigmoid(x):
    return 1/(1+np.exp(-x))
x = np.arange(-5.0, 5.0, 0.1)
y1 = sigmoid(x+0.5)
y2 = sigmoid(x+1)
y3 = sigmoid(x+1.5)

plt.plot(x, y1, 'r', linestyle='--') # x + 0.5
plt.plot(x, y2, 'g') # x + 1
plt.plot(x, y3, 'b', linestyle='--') # x + 1.5
plt.plot([0,0],[1.0,0.0], ':') # 가운데 점선 추가
plt.title('Sigmoid Function')
plt.show()
```

시그모이드 함수는 입력값이 커지면 1에 수렴하고, 작아지면 0에 수렴.

0부터 1까지의 값에서 출력값이 0.5보다 작으면 0(False), 크면 1(True)로 판단해서 사용하면 이진 분류 문제로 사용 가능
